```python
import re

final_lrc = []
# re 正規表達
# 中文（簡/繁體）在unicode中是u4e00~u9fa5
zhPattern = re.compile(u'[\u4e00-\u9fa5]+')

# 遍历所有歌词行
for line in lrc_lines:
   # findall会返回一个列表
   zh_list = zhPattern.findall(line)
   # 将中文拷贝到 final_lrc 中去
   for item in zh_list:
        final_lrc.append(item + '\n')
        
# 这样就只剩中文啦！
print(final_lrc)
print(len(final_lrc))
```

```python
# 字典
word_to_ix = {}

for line in final_lrc:
    for word in line:
        if word not in word_to_ix:
        # 在單詞表的末端添加這個單詞
            word_to_ix[word] = len(word_to_ix)

# 打印出字典中的前10個元素，注意它是無序的
for i, key in zip(range(10), word_to_ix):
    print(key, ":", word_to_ix[key])
```

```python
import operator
# 排序
sorted_char_list = sorted(word_to_ix.items(), key=operator.itemgetter(1), reverse=False)

# 有序列表
# 這裡注意字典是「dict」類型
# 有序列表是「list」類型
char_list = []
for item in sorted_char_list:
    char_list.append(item[0])

# 有序列表中的元素
print(char_list[:18])

# 觀察有序列表中的元素即字典的有序排列
for ch in char_list[:18]:
    print(ch, ":", word_to_ix[ch])
```

```python
import random

chunk_len = 200

# 隨機選取200 + 1個字符的數據
def random_chunk():
    # 起點的可選範圍：0 ~ final_lrc_len-chunk_len
    start_index = random.randrange(0, len(big_string) - chunk_len)
    end_index = start_index + chunk_len + 1
    # 將抽取的列表項轉化為字符串再返回
    return big_string[start_index:end_index]

result = random_chunk()
print(len(result))
result
```

```python
import torch
import torch.nn as nn
from torch.autograd import Variable

def char_tensor(string):
    # 先按指定長度創建一個longTensor，填充0
    tensor = torch.zeros(len(string)).long()
    # 逐字查找字典
    # 取出每個字的索引號，保存到 Tensor 中
    for c in range(len(string)):
        tensor[c] = word_to_ix[string[c]]
    return Variable(tensor)

# 來試試這個方法
index_tensor = char_tensor(random_chunk())

# 查看 Tensor 長度
print(len(index_tensor))
# 查看 Tensor 中保存的索引
print(index_tensor[:10])
```

```python
def random_training_set():
    # 先選取201字符長度的文本
    chunk = random_chunk()
    # 將選取的文本全部轉化為 Tensor
    inp = char_tensor(chunk[:-1])
    target = char_tensor(chunk[1:])
    return inp, target

inp, target = random_training_set()
print(inp)
print(len(target))
```

```python
import torch
import torch.nn as nn
from torch.autograd import Variable
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, n_layers):
        super(RNN, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.n_layers = n_layers
        # 編碼器
        self.encoder = nn.Embedding(input_size, hidden_size)
        # GRU
        self.gru = nn.GRU(hidden_size, hidden_size, n_layers)
        # 解碼器
        self.decoder = nn.Linear(hidden_size, output_size)
    def forward(self, input, hidden):
        input = self.encoder(input.view(1, -1))

        output, hidden = self.gru(input.view(1, 1, -1), hidden)
        output = self.decoder(output.view(1, -1))
        return output, hidden
    def init_hidden(self):
        return Variable(torch.zeros(self.n_layers, 1, self.hidden_size))
```

```python
n_epochs = 3500 # 迭代次數
print_every = 1 # 每隔多少次迭代打印
plot_every = 10 # 每個多少次迭代記錄圖表值
hidden_size = 100 # 隱藏單元的數量
n_layers = 1 # 隱藏層層數
lr = 0.005 # 學習率
# 實例化RNN模型為decoder
decoder = RNN(n_characters, hidden_size, n_characters, n_layers)
# 本次使用 Adam 優化算法
decoder_optimizer = torch.optim.Adam(decoder.parameters(), lr=lr)
# 使用交叉熵損失
criterion = nn.CrossEntropyLoss()
```

```python
# inp，訓練輸入數據，包含200個字
# target，目標，包含200個字
def train(inp, target):
    # 將隱藏層置零
    hidden = decoder.init_hidden()
    # 將模型的梯度置零
    decoder.zero_grad()
    # 損失值初始化
    loss = 0
    # chunk_len是在上節定義的訓練數據及目標數據的長度
    # 值為200
    for c in range(chunk_len):
        # 調用RNN的forward函數，進行一次正向傳播
        output, hidden = decoder(inp[c], hidden)
        # 與目標值相比計算交叉熵損失
        loss += criterion(output, target[c].unsqueeze(0))
    # 循環完200個字後進行一次反向傳播
    loss.backward()
    decoder_optimizer.step()
    # 返回平均損失
    return loss.item() / chunk_len
```

```python
# prime_str：用戶給的「開頭」
# predict_len：要預測多少個字，每次產生歌詞的長度
# temperature：混亂度
def evaluate(prime_str='我', predict_len=100, temperature=0.8):
    # 隱藏層置零
    hidden = decoder.init_hidden()
    # 將我們給模型的「開頭」轉化為 Tensor
    prime_input = char_tensor(prime_str)
    # predicted保存生成的嘻哈歌詞
    predicted = prime_str
    # 首先用用戶給的「開頭」運行模型的forward
    # 也就是使用「開頭」建立RNN隱藏層狀態
    for p in range(len(prime_str) - 1):
        _, hidden = decoder(prime_input[p], hidden)
    inp = prime_input[-1]
    for p in range(predict_len):
        # 输出以及隐藏层状态
#        print('inp.shape', inp.shape, predict_len)
        output, hidden = decoder(inp, hidden)
#        print('inp.shape1', inp.shape, predict_len)
        # 根据“混乱度”从输出的多项式分布中采样
        output_dist = output.data.view(-1).div(temperature).exp()
#        print('inp.shape3', inp.shape, predict_len)
        top_i = torch.multinomial(output_dist, 1)[0]
#        print('inp.shape4', inp.shape, predict_len, top_i)
        # 将采样的值转化为文字
        predicted_char = char_list[top_i]
        print('inp.shape5', inp.shape, predicted_char)
        # 将模型预测出的字加入到predicted中
        predicted += predicted_char
        print('inp.shape6', inp.shape, predict_len, predicted, predicted_char)
        # 下一次预测的“输入”就是这次预测的输出
        inp = char_tensor(predicted_char)       
        print('inp.shape7', inp.shape,inp,  char_tensor(predicted_char))
    return predicted
```

```python
import time, math
# 用於計算訓練時間的函數
def time_since(since):
    s = time.time() - since
    m = math.floor(s / 60)
    s -= m * 60
    return '%dm %ds' % (m, s)
```

```python
# 開始時間
start = time.time()
all_losses = []
loss_avg = 0
for epoch in range(1, n_epochs + 1):
    loss = train(*random_training_set())
    loss_avg += loss
    # 每個一段時間，讓模型生成一次看看效果
    if epoch % print_every == 0:
        print('[%s (%d %d%%) %.4f]' % (time_since(start), epoch, epoch / n_epochs * 100, loss))
        print(evaluate('我', 100), '')
    if epoch % plot_every == 0:
        all_losses.append(loss_avg / plot_every)
        loss_avg = 0
```
