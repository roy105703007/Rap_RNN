```python
import urllib.request as req
import time
import bs4 
f = open('/Users/Jerry/Downloads/song_list_file3.txt',"r")    

lines = f.readlines()

print(len(lines))

song_url_list=[line[:-1] for line in lines]

f.close()

file = open('/Users/Jerry/Desktop/lyrics.txt', 'a')

for i in range(len(song_url_list)):
    time.sleep(3)
    url = song_url_list[i]
    request=req.Request(url, headers={
    'User-Agent':"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36"})
    
    with req.urlopen(request) as response:
        data = response.read().decode("utf-8")
    root = bs4.BeautifulSoup(data, "html.parser")

    titles = root.find("div", class_ = "song-content")
    
    lyrics = str(titles.p)
    
    file.write(lyrics + "\n")
    
    print('suc \n\n')

file.close()
```
